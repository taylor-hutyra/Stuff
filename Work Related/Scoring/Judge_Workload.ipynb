```python
from tkinter import *
from setup import *
from random import random
import pandas as pd
import numpy as np
import glob
import os

###################################GLOBALS
CURRENT_SESSION = 3
color="purple"
default_color="white"
ROWS = 8
COLUMNS = 6
NUMJUDGES = 4
#############################################

device_list = pd.read_csv('judge_devices.csv')
display(device_list)
df = pd.read_csv("gb_id_projects.csv")
display(df)
workload_list = []
workload_list.append((pd.read_csv('ses1_workload.csv'))[['judge id','work']])

workload_list.append((pd.read_csv('ses2_workload.csv'))[['judge id','work']])
workload_list.append((pd.read_csv('ses3_workload.csv'))[['judge id','work']])
workload_list.append((pd.read_csv('ses4_workload.csv'))[['judge id','count']])


for x in workload_list:
    x.columns = ['RESPONSE ID', 'WORKLOAD']
display(workload_list[1])
# display(df)
pdf = df[['session', 'location','PRESENTATION TYPE','NAME','GUIDEBOOK ID']]
pdf = pdf.loc[pdf['PRESENTATION TYPE']=='poster']

poster_sessions = []
poster_list_per_session =[ ]
for i in range(4):
    tf = pdf[pdf['session'] == i+1]
    poster_sessions.append(tf.sort_values('location').reset_index(drop=True))
    poster_list_per_session.append(list(tf['GUIDEBOOK ID']))

# display(poster_sessions[1])
# display((poster_sessions[3].loc[poster_sessions[3]['GUIDEBOOK ID']=='8786706'])['location'])

poster_list = poster_sessions[CURRENT_SESSION-1]
# print(poster_list)
poster_num =len(poster_list)
session_poster_list = list(poster_list['GUIDEBOOK ID'])
# print(session_poster_list)
print('Num Posters: ' + str(poster_num))
print('Rows: ' + str(ROWS) + ', Cols: ' + str(COLUMNS))
# display(poster_list)


list_of_files = glob.glob('./session_update/*') # * means all if need specific format then *.csv
latest_file = max(list_of_files, key=os.path.getctime)
uf = pd.read_csv(latest_file, names = ['Judge_ID','Poster_ID','Presenter', 'Question_ID', 'Question', 'Response','Date'])
nuf = uf.loc[uf['Poster_ID'].isin(list(poster_list['GUIDEBOOK ID']))]
# display(nuf)
nuf = nuf.loc[nuf['Question']=='Judges Password']
# display(nuf)
nuf = nuf.loc[nuf['Response'].str.lower()=='score']

nuf = nuf[['Judge_ID','Poster_ID',]]
device_list = device_list[['DEVICE ID','RESPONSE ID','NAME', 'SUBFIELD','CATEGORY','INSTITUTION']]
display(nuf)
print('DEVICE LIST HERE')
display(device_list)

judge_counts = nuf.groupby('Judge_ID').apply(lambda g: len(g['Poster_ID'].unique())).reset_index()
judge_counts.columns = ['DEVICE ID', 'COUNTS']
display(judge_counts)

merge_one = pd.merge(judge_counts, device_list, on = 'DEVICE ID', how = 'outer')
display(merge_one)
merge_two = pd.merge(merge_one, workload_list[CURRENT_SESSION-1], on = 'RESPONSE ID')
merge_two = merge_two[['NAME', 'COUNTS','WORKLOAD','SUBFIELD','CATEGORY','INSTITUTION']]
display(merge_two)

print_df = merge_two[['NAME', 'COUNTS','WORKLOAD','SUBFIELD','CATEGORY','INSTITUTION']]
print_df.to_html('session_'+str(CURRENT_SESSION)+'_workload.html')
```
