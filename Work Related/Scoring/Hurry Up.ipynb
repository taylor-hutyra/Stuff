```python
from tkinter import *
from setup import *
from random import random
import pandas as pd
import numpy as np
import glob
import os

df = pd.read_csv("gb_id_projects.csv")

# list_of_files = glob.glob('./session_update/*') # * means all if need specific format then *.csv
# latest_file = max(list_of_files, key=os.path.getctime)
# pf = pd.read_csv(latest_file, names = ['Judge_ID','Poster_ID','Presenter', 'Question_ID', 'Question', 'Response','Date']).drop(0)
# pf = pd.read_csv('orals.csv', names = ['Judge_ID','Poster_ID','Presenter', 'Question_ID', 'Question', 'Response','Date']).drop(0)
pf = pd.read_csv('last_guidebook.csv', names = ['Judge_ID','Poster_ID','Presenter', 'Question_ID', 'Question', 'Response','Date']).drop(0)

# display(pf)

uf = pf.groupby('Poster_ID').apply(lambda g: g.loc[g['Question']=='Feedback/Comments']['Response'].reset_index()).reset_index()
uf.columns = ['GUIDEBOOK ID', 'Judge', 'Thing','Feedback']
uf.drop('Thing', axis = 1 ,inplace=True)
# display(uf)

newdf = pd.merge(uf, df, on='GUIDEBOOK ID')
print(newdf.columns)
display(newdf.loc[newdf['PRESENTATION TYPE'].str.lower() == 'poster'])
# print(newdf.columns)
first_df = newdf[['GUIDEBOOK ID','FIRST NAME','LAST NAME','INSTITUTION','EMAIL 1', 'FACULTY ADVISOR', 'FACULTY ADVISOR EMAIL','Feedback']]
display(first_df)
def new_f(grp):
    thisdf = grp['Feedback'].reset_index(drop=True).to_frame().unstack().unstack()
    return(thisdf)
feedback_df = newdf.groupby('GUIDEBOOK ID').apply(new_f)#.drop(0, axis=1)
feedback_df.index = feedback_df.index.droplevel(1)
feedback_df.reset_index(inplace = True)
print(feedback_df.index)
display(feedback_df)
feedback_df.columns = ['GUIDEBOOK ID','Feedback 1','Feedback 2','Feedback 3','Feedback 4','Feedback 5', 'Feedback 6']
display(feedback_df)

final_df = pd.merge(feedback_df,first_df, on ='GUIDEBOOK ID', how = 'left').drop('Feedback',axis=1).drop_duplicates().reset_index(drop=True)
display(final_df)
# final_df.to_csv('oral_email_merge.csv')
```
