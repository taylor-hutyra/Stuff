```python
from setup import *
df = pd.read_csv('guidebook_score_export.csv')

#Check for invalid non-numeric scores
idx = df['Response'].str.isnumeric()
print("%u valid scores and %u invalid scores (listed below)"%(idx.sum(),(~idx).sum()))
#display(df[~idx])
df['Response'] = pd.to_numeric(df['Response'],errors='coerce')

#Exploratory data analysis
# for col in df:
#     display(df[col].value_counts())
# Check that 2 pairs of columns are redundant
# df.groupby('Question ID')['Question Text'].apply(pd.value_counts)
# df.groupby('Session/Item ID')['Session/Item Name'].apply(pd.value_counts)




#Convert judge raw scores to z-scores.  This normalizes judges to adjust
#for "easy" judges and "hard" judges.  For each judge, compute that judge's mean
#and standard deviation.  For of her scores, subtract her mean and divide by her standard deviation."

R = df.groupby('Device ID')['Response']
print("Summary stats for judges before normalization")
display(R.describe().unstack())

norm_df = df.copy()
norm_df['Response'] = R.transform(lambda x: (x-x.mean())/x.std())
print("Summary stats for judges before after normalization.  Note evvery mean=0 and std=1, as desired.")
display(norm_df.groupby('Device ID')['Response'].describe().unstack())
print("Rankings of participants using standardized scores")
display(norm_df.groupby('Session/Item Name')['Response'].describe().unstack().sort_values(by='mean',ascending=False))
```
